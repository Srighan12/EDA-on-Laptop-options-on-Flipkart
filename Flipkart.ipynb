```python
# To get the top 5 most rated laptop

top5 = df.sort_values('Ratings',ascending = False).head(5)
top5
```

```python
#plot to show the Top 5 rated laptops

plt.figure(figsize=(15,7))
plt.bar(x=top5["ProductName"],height=top5["Ratings"])
plt.title("Top 5 rated Laptops",fontsize=15)
plt.xlabel("Product Name",fontsize=15)
plt.xticks(rotation=270)
plt.ylabel("Ratings",fontsize=15)
plt.show()
```

```python
df. sort_values(['Current_Price', 'Stars'], ascending=[True, False]).head()
```

```python
# Best laptop for school students
# considering budget to be between 15000 to 25000
df_school = df[(df['Current_Price'] > 15000) & (df['Current_Price'] < 25000)].sort_values('Ratings', ascending=False).head(10).sort_values('Stars', ascending=False).head(5)
df_school
```

```python
# from this we can get to know that ASUS Chromebook Flip Touch Celeron Dual Core is the best laptop for school students 
# considering the budget ,Rating , Discounted price and specifications
```

```python
# Best laptop for college students
# considering budget to be between 25000 to 50000
df_college = df[(df['Current_Price'] > 25000) & (df['Current_Price'] < 50000)].sort_values('Ratings', ascending=False).head(10).sort_values('Stars', ascending=False).head(5)
df_college
```

```python
# from this we can get to know that ASUS Core i5 10th Gen is the best laptop for college students 
# considering the budget ,Rating , Discounted price and specifications
```

```python
# Best laptop for working professionals
# considering budget to be between 50000 to 100000
df_working = df[(df['Current_Price'] > 50000) & (df['Current_Price'] < 100000)].sort_values('Ratings', ascending=False).head(10).sort_values('Stars', ascending=False).head(5)
df_working
```

```python
# from this we can get to know that HP Pavilion Ryzen 5 Hexa Core AMD R5 is the best laptop for working professionals 
# considering the budget ,Rating , Discounted price and specifications
```

```python
# most reviewed and rated laptop

df_review = df.sort_values('Reviews',ascending = False).head(5)
df_review
```

```python
# To find the value for money laptop
# Value for money would be the best mid range laptop
midrange_laptop = df[(df["Current_Price"] > 35000) & (df["Current_Price"] < 45000)].sort_values('Stars',ascending = False).head()
midrange_laptop
```
